```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/Chicago_Crimes_2012_to_2017.csv"
chicago_df = pd.read_csv(csv_file)

chicago_df.head()
```

```python
chicago_df = chicago_df.drop(columns=['Unnamed: 0','ID', 'Case Number', 'Location', 'Updated On', 'IUCR'])
```

```python
chicago_df
```

```python
chicago_df.drop(chicago_df.index[chicago_df['Year']>2016], inplace = True)

chicago_df.drop(chicago_df.index[chicago_df['Year']<2013], inplace = True)
```

```python
chicago_df.head()
```

```python
chicago_df.columns
```

```python
chicago_df['Datetime'] = chicago_df['Date'].astype('datetime64')
```

```python
chicago_df.dtypes
```

```python
chicago_df
```

```python
connection_string = "postgres:postgres@localhost:5432/etl_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
chicago_df.to_sql(name='chicago_crime', con=engine, if_exists='append', index=True)
```
